MANAGING FOLDER, moves files from Datadump folder to Datadumparchive if file creation date is before target date

```python
import os
import shutil
import glob
import datetime
#source_folder
#destination_folder
source_folder = "C:/Users/99111/OneDrive/Documents/pynse/Shashank work/data/raw/Data Dump"
destination_folder = "C:/Users/99111/OneDrive/Documents/pynse/Shashank work/data/raw/Data Dumparchive"
target_date = datetime.datetime(2024, 1, 3)  # Change this to your desired date

# Get a list of all files in the source folder
file_list = glob.glob(os.path.join(source_folder, '*'))

# Move the files to the destination folder based on the target date
files_moved = 0
for file_path in file_list:
    file_name = os.path.basename(file_path)
    modified_time = datetime.datetime.fromtimestamp(os.path.getmtime(file_path))
    
    if modified_time <= target_date:
        destination_path = os.path.join(destination_folder, file_name)
        shutil.move(file_path, destination_path)
        files_moved += 1

print(f"{files_moved} files moved to the destination folder.")
```
