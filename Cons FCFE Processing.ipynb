```python
# consolidated net profit

folder_path = "C:/Users/99111/OneDrive/Desktop/Full data set/Financial Reports/PNL Cons/*.csv"

files = glob.glob(folder_path)

df_list = []

for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

PNL = pd.concat(df_list)

PNL['PLC_Year End'] = pd.to_datetime(PNL['PLC_Year End'], format='%Y%m')

PNL['PLC_Year End'] = PNL['PLC_Year End'] + pd.offsets.MonthEnd(1)

PNL = PNL.sort_values('PLC_Year End')

# Cash flow statement

folder_path = "C:/Users/99111/OneDrive/Desktop/Full data set/Financial Reports/Cashflow Statement Cons/*.csv"

files = glob.glob(folder_path)

df_list = []

for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

CFS = pd.concat(df_list)

CFS = pd.concat(df_list)

CFS['CFC_Year End'] = pd.to_datetime(CFS['CFC_Year End'], format='%Y%m')

CFS['CFC_Year End'] = CFS['CFC_Year End'] + pd.offsets.MonthEnd(1)

CFS = CFS.sort_values('CFC_Year End')

# Balance Sheet

folder_path = "C:/Users/99111/OneDrive/Desktop/Full data set/Financial Reports/Balance Sheet Cons/*.csv"

files = glob.glob(folder_path)

df_list = []

for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

BS = pd.concat(df_list)

BS['BSC_Year End'] = pd.to_datetime(BS['BSC_Year End'], format='%Y%m')

BS['BSC_Year End'] = BS['BSC_Year End'] + pd.offsets.MonthEnd(1)    


u1 = PNL['Company Name'].unique().tolist()
u2 = CFS['Company Name'].unique().tolist()
u3 = BS['Company Name'].unique().tolist()
uniques = list(set(u1).intersection(u2).intersection(u3))

```

BS, CF and PNL  and Calcs

```python
company_names = uniques

# Step 2: Iterate through the list of Company names using a loop
results1 = pd.DataFrame()
for company_name in company_names:
    # Execute the code for each Company
    PNL1 = PNL[PNL['Company Name'] == company_name]
    PNL1 = PNL1.sort_values('PLC_Year End', ascending=True)
    PNL1.fillna(0, inplace=True)
    PNL1 = PNL1.rename(columns={'PLC_Year End': 'Date'})
    PNL1['uniquecode'] = PNL1['Company Name'] + PNL1['Date'].astype(str)
    PNL1.drop_duplicates(subset='uniquecode', keep='last', inplace=True)
    PNL1 = PNL1.reset_index(drop=True)

    PNL1['Other Income / Operating Income'] = (PNL1['PLC_Other Income'] / PNL1['PLC_Operating Profit (Excl OI)'] ) * 100

    PNL1['Misc & Capitalised Expense'] = PNL1['PLC_Miscellaneous Expenses'] + PNL1['PLC_Less: Expenses Capitalised'] 
    PNL1['change in Misc & Capitalised Expense'] = PNL1['Misc & Capitalised Expense'] - PNL1['Misc & Capitalised Expense'].shift(1)
    PNL1['Misc & Capitalised Expense / Total Expenditure'] = (PNL1['Misc & Capitalised Expense'] /  PNL1['PLC_Total Expenditure']) * 100 


    PNL1['Net Sales Growth'] = ((PNL1['PLC_Net Sales'] -  PNL1['PLC_Net Sales'].shift(1)) / PNL1['PLC_Net Sales'].shift(1)) * 100
    PNL1['Operating Profit Growth'] = ((PNL1['PLC_Operating Profit (Excl OI)'] -  PNL1['PLC_Operating Profit (Excl OI)'].shift(1)) / PNL1['PLC_Operating Profit (Excl OI)'].shift(1)) * 100 
    PNL1['PAT Growth'] = ((PNL1['PLC_Profit After Tax'] -  PNL1['PLC_Profit After Tax'].shift(1)) / PNL1['PLC_Profit After Tax'].shift(1)) * 100

    PNL1['Minority share / PAT'] = (PNL1['PLC_Minority Interest'] / PNL1['PLC_Profit After Tax'] ) * 100
    
    # Step 3: Append the results of each Company to a single dataframe
    PNL1['Company Name'] = company_name
    results1 = pd.concat([results1, PNL1])


results2 = pd.DataFrame()
for company_name in company_names:
    # Execute the code for each Company
    CFS1 = CFS[CFS['Company Name'] == company_name]
    CFS1 = CFS1.sort_values('CFC_Year End', ascending=True)
    CFS1.fillna(0, inplace=True)
    CFS1 = CFS1.rename(columns={'CFC_Year End': 'Date'}) 
    CFS1['uniquecode'] = CFS1['Company Name'] + CFS1['Date'].astype(str)
    CFS1.drop_duplicates(subset='uniquecode', keep='last', inplace=True)
    CFS1 = CFS1.reset_index(drop=True) 

    CFS1['CAPEX'] = -1 * (CFS1['CFC_Purchase of Fixed Assets'] + CFS1['CFC_Investment in Subsidiaries'] + CFS1['CFC_Advances for capital expenditure'] + CFS1['CFC_Sale of Fixed Assets']) - CFS1['CFC_Depreciation'] - CFS1['CFC_Impairment']

    CFS1['Net Casflow from Debt'] = CFS1['CFC_Repayment of Long Term Borrowings'] + CFS1['CFC_Repayment of Debenture / Bonds'] + CFS1['CFC_Proceeds from Long Term Borrowings'] + CFS1['CFC_Proceeds from Debenture / Bonds']

    CFS1['Actual payout to Equity'] = -1 * (CFS1['CFC_Equity Dividend Paid'] + CFS1['CFC_Buy Back of Equity Shares Capital'])

    CFS1['Funny Items in operating CFs'] = CFS1['CFC_Bad debts irrecoverable written off'] + CFS1['CFC_Provision for doubtful debts & advances'] + CFS1['CFC_Misc. Expenses written off'] + CFS1['CFC_Other Adjustments']
    CFS1['change in Funny Items in operating CFs'] = CFS1['Funny Items in operating CFs'] - CFS1['Funny Items in operating CFs'].shift(1)

    CFS1['NetfromPreferencialEquity'] = CFS1['CFC_Proceeds from Preference Shares Capital'] + CFS1['CFC_Redemption of Preference Shares Capital'] + CFS1['CFC_Preference Dividend']


    # Step 3: Append the results of each Company to a single dataframe
    CFS1['Company Name'] = company_name
    results2 = pd.concat([results2, CFS1])


results3 = pd.DataFrame()
for company_name in company_names:
    # Execute the code for each Company
    BS1 = BS[BS['Company Name'] == company_name]
    BS1 = BS1.sort_values('BSC_Year End', ascending=True)
    BS1.fillna(0, inplace=True)
    BS1 = BS1.rename(columns={'BSC_Year End': 'Date'})
    BS1['uniquecode'] = BS1['Company Name'] + BS1['Date'].astype(str)
    BS1.drop_duplicates(subset='uniquecode', keep='last', inplace=True)
    BS1 = BS1.reset_index(drop=True)

    BS1['Minority interest  / Total NC Liability'] = (BS1['BSC_Minority Interest'] / BS1['BSC_Total Non-Current Liabilities']) * 100

    BS1['LT Payables & provisions & other'] = BS1['BSC_Long Term Trade Payables'] + BS1['BSC_Long Term Provisions'] + BS1['BSC_Other Long Term Liabilities']
    BS1['Change in LT Payables & provisions & other'] = BS1['LT Payables & provisions & other'] - BS1['LT Payables & provisions & other'].shift(1)
    BS1['LT Payables & provisions / Total NC liabilities'] = (BS1['LT Payables & provisions & other'] / BS1['BSC_Total Non-Current Liabilities']) * 100

    BS1['Change in Net Block'] = (BS1['BSC_Net Block'] - BS1['BSC_Net Block'].shift(1)) * 100

    BS1['FunnyAssets'] = BS1['BSC_Capital Work in Progress'] + BS1['BSC_Intangible assets under development'] + BS1['BSC_Pre-operative Expenses pending'] + BS1['BSC_Assets in transit']
    BS1['Change in FunnyAssets'] = BS1['FunnyAssets'] - BS1['FunnyAssets'].shift(1)
    BS1['FunnyAssets Flag'] = (BS1['FunnyAssets'] / BS1['BSC_Total Non-Current Assets']) * 100

    BS1['Contingent Liability / Total Assets'] = BS1['BSC_Contingent Liabilities'] / BS1['BSC_Total Assets']
    BS1['Contingent liability increase Flag'] = BS1['BSC_Contingent Liabilities'] - BS1['BSC_Contingent Liabilities'].shift(1)

    BS1['Non Cash Working Capital'] = BS1['BSC_Total Current Assets Excluding Current Investments'] - BS1['BSC_Total Current Liabilities'] - BS1['BSC_Cash and Bank']
    BS1['Change in Non Cash Working Capital'] = BS1['Non Cash Working Capital'] - BS1['Non Cash Working Capital'].shift(1)
    BS1['Cash/Current Liability'] = (BS1['BSC_Cash and Bank'] / BS1['BSC_Total Current Liabilities'])*100

    # Step 3: Append the results of each Company to a single dataframe
    BS1['Company Name'] = company_name
    results3 = pd.concat([results3, BS1])

merged_df = pd.merge(results1, results2, on='uniquecode', how='outer')
merged_df = pd.merge(merged_df, results3, on='uniquecode', how='outer')
merged_df['FCFE'] = merged_df['PLC_Consolidated Net Profit'] - merged_df['CAPEX'] + merged_df['Net Casflow from Debt'] - merged_df['Change in Non Cash Working Capital'] + merged_df['NetfromPreferencialEquity']


merged_df = merged_df[['Company Name_x', 'CD_Trading Status_x', 'CD_BSE Code_x', 'CD_NSE Symbol_x', 'Date_x','uniquecode' ,'PLC_Net Sales', 'PLC_Miscellaneous Expenses', 'PLC_Less: Expenses Capitalised', 'PLC_Total Expenditure', 'PLC_Operating Profit (Excl OI)', 'PLC_Other Income', 'PLC_Operating Profit', 'PLC_Profit After Tax', 'PLC_Minority Interest',  'CFC_Depreciation', 'CFC_Impairment', 'CFC_Bad debts irrecoverable written off', 'CFC_Provision for doubtful debts & advances', 'CFC_Misc. Expenses written off', 'CFC_Other Adjustments', 'CFC_Purchase of Fixed Assets', 'CFC_Sale of Fixed Assets', 'CFC_Investment in Subsidiaries', 'CFC_Advances for capital expenditure', 'CFC_Proceeds from Long Term Borrowings', 'CFC_Repayment of Long Term Borrowings', 'CFC_Proceeds from Debenture / Bonds', 'CFC_Repayment of Debenture / Bonds', 'CFC_Proceeds from Preference Shares Capital', 'CFC_Redemption of Preference Shares Capital','CFC_Buy Back of Equity Shares Capital','CFC_Equity Dividend Paid','CFC_Preference Dividend','CFC_Loans (to) / from subsidiaries','BSC_Share Capital','BSC_Total Reserves','BSC_Minority Interest','BSC_Other Long Term Liabilities','BSC_Long Term Trade Payables','BSC_Long Term Provisions','BSC_Total Non-Current Liabilities','BSC_Total Current Liabilities','BSC_Net Block','BSC_Capital Work in Progress','BSC_Intangible assets under development','BSC_Pre-operative Expenses pending','BSC_Assets in transit','BSC_Total Non-Current Assets','BSC_Inventories','BSC_Cash and Bank','BSC_Total Current Assets Excluding Current Investments','BSC_Total Assets','BSC_Contingent Liabilities','BSC_Total Debt','BSC_Book Value','BSC_Adjusted Book Value','Other Income / Operating Income', 'Misc & Capitalised Expense', 'change in Misc & Capitalised Expense', 'Misc & Capitalised Expense / Total Expenditure', 'Net Sales Growth', 'Operating Profit Growth', 'PAT Growth', 'Minority share / PAT',  'Actual payout to Equity', 'Funny Items in operating CFs', 'change in Funny Items in operating CFs',  'Minority interest  / Total NC Liability', 'LT Payables & provisions & other', 'Change in LT Payables & provisions & other', 'LT Payables & provisions / Total NC liabilities', 'Change in Net Block', 'FunnyAssets', 'Change in FunnyAssets', 'FunnyAssets Flag', 'Contingent Liability / Total Assets', 'Contingent liability increase Flag', 'Non Cash Working Capital',  'Cash/Current Liability','PLC_Consolidated Net Profit','CAPEX','Net Casflow from Debt','NetfromPreferencialEquity', 'Change in Non Cash Working Capital','FCFE']]

merged_df['PLC_Net Sales_4y'] = merged_df['PLC_Net Sales'].rolling(window=4).sum()
merged_df['PLC_Operating Profit_4y'] = merged_df['PLC_Operating Profit'].rolling(window=4).sum()
merged_df['PLC_Consolidated Net Profit_4y'] = merged_df['PLC_Consolidated Net Profit'].rolling(window=4).sum()
merged_df['CAPEX_4y'] = merged_df['CAPEX'].rolling(window=4).sum()
merged_df['Net Casflow from Debt_4y'] = merged_df['Net Casflow from Debt'].rolling(window=4).sum()
merged_df['NetfromPreferencialEquity_4y'] = merged_df['NetfromPreferencialEquity'].rolling(window=4).sum()
merged_df['Change in Non Cash Working Capital_4y'] = merged_df['Change in Non Cash Working Capital'].rolling(window=4).sum()
merged_df['FCFE_4y'] = merged_df['FCFE'].rolling(window=4).sum()


filepath = r'C:\Users\99111\OneDrive\Desktop\Full data set\FCFE\Data\combinedfile.csv'
merged_df.to_csv(filepath, index=False)
```
